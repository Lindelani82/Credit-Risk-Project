# Data Preprocessing

```python
def impute_missing_values(df):
    """
    Imputes missing values in a DataFrame.
    - Numerical columns: Fill with the mean.
    - Categorical/object columns: Fill with the mode.

    Parameters:
    df (pd.DataFrame): Input DataFrame.

    Returns:
    pd.DataFrame: DataFrame with missing values imputed.
    """
    
    # Iterate through all columns
    for column in df.columns:
        if df[column].isnull().any():  # Check if the column has missing values
            
            # For numerical columns
            if pd.api.types.is_numeric_dtype(df[column]):  
                df[column].fillna(df[column].mean(), inplace=True)
            
            # For categorical/object columns
            elif pd.api.types.is_categorical_dtype(df[column]) or df[column].dtype == 'object':
                if not df[column].mode().empty:  # Check if a mode exists
                    df[column].fillna(df[column].mode()[0], inplace=True)

    return df
```

```python
new_data = impute_missing_values(new_data)
new_data.isnull().sum()
```

# Statistics

Objective: Determine if person_income (numerical) impacts cb_person_default_on_file (categorical: 'Y' and 'N').



**Null Hypothesis (H0):**
There is no significant difference in person_income between those who defaulted ('Y') and those who did not default ('N').

**Alternative Hypothesis (Ha):**

There is a significant difference in person_income between those who defaulted ('Y') and those who did not default ('N').

*Test to use:* Independent t-test (for two independent groups: 'Y' and 'N').

```python
from scipy.stats import ttest_ind

# Separate groups based on the target column
default = new_data[new_data['cb_person_default_on_file'] == 'Y']['person_income']
non_default = new_data[new_data['cb_person_default_on_file'] == 'N']['person_income']

# Perform t-test
t_stat, p_val = ttest_ind(default, non_default, equal_var=False)
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

**P-value Analysis:**

The p-value (0.4872) is greater than 0.05, meaning we fail to reject the null hypothesis.


- There is no significant difference in person_income between those who defaulted ('Y') and those who did not default ('N') at the 5% significance level.
- This suggests that person_income does not have a significant impact on whether a person defaults or not, based on this dataset.

###  Hypothesis Testing for Categorical vs. Categorical Relationship

Objective: Test if loan_grade (categorical) is associated with cb_person_default_on_file (categorical: 'Y' and 'N').

**Null Hypothesis (H0):**
There is no association between loan_grade and cb_person_default_on_file.

**Alternative Hypothesis (Ha):**
There is an association between loan_grade and cb_person_default_on_file.

**Test to use:** Chi-Square Test for Independence.

```python
# Create a contingency table
contingency_table = pd.crosstab(new_data['loan_grade'], new_data['cb_person_default_on_file'])

# Perform Chi-Square test
chi2, p_val, dof, expected = chi2_contingency(contingency_table)
print(f"Chi2: {chi2}, P-value: {p_val}")
```

**Interpretation of the Results:**

**P-value Analysis:**

The p-value (0.0) is much less than 0.05, meaning we reject the null hypothesis.

**Conclusion:**

- There is a significant association between loan_grade and cb_person_default_on_file.
- This suggests that the distribution of loan grades is not independent of whether a person defaults or not. In simpler terms, loan_grade plays an important role in predicting the likelihood of default.

```python


# Create a contingency table for loan_grade and cb_person_default_on_file
contingency_table = pd.crosstab(new_data['loan_grade'], new_data['cb_person_default_on_file'])

# Plot the stacked bar chart
contingency_table.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title("Stacked Bar Chart of Loan Grade vs Default Status")
plt.xlabel("Loan Grade")
plt.ylabel("Count")
plt.legend(title="Default Status (Y/N)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Loan Amount Distribution 

**Null Hypothesis (H0):** Loan amount (loan_amnt) is normally distributed.

**Alternative Hypothesis (Ha):** Loan amount is not normally distributed.

```python
# Plotting a histogram with KDE
plt.figure(figsize=(10, 6))
sns.histplot(new_data['loan_amnt'], kde=True, color='blue', bins=30, alpha=0.6)
plt.title('Loan Amount Distribution (Histogram + KDE)')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

from scipy.stats import shapiro

# Shapiro-Wilk test for normality
stat, p_val = shapiro(new_data['loan_amnt'])
print(f"Shapiro-Wilk Test Statistic: {stat}, P-value: {p_val}")
```

Since the p-value is 0.0, which is less than the common significance level (e.g., 0.05), we reject the null hypothesis. This indicates that the loan amount is not normally distributed.

```python
from scipy.stats import probplot

# Q-Q Plot to assess normality
plt.figure(figsize=(8, 6))
probplot(new_data['loan_amnt'], dist="norm", plot=plt)
plt.title('Q-Q Plot for Loan Amount')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Feature Engineering

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

def preprocess_categorical_data(df, label_encode_cols, one_hot_encode_cols):
    """
    Function to label encode specified columns and one-hot encode others.

    Parameters:
    - df (pd.DataFrame): Input DataFrame.
    - label_encode_cols (list): Columns to label encode.
    - one_hot_encode_cols (list): Columns to one-hot encode.

    Returns:
    - pd.DataFrame: Preprocessed DataFrame with categorical columns encoded.
    """
    # Create a copy to avoid modifying the original DataFrame
    df_processed = df.copy()

    # Label Encode specified columns
    label_encoder = LabelEncoder()
    for col in label_encode_cols:
        df_processed[col] = label_encoder.fit_transform(df_processed[col])
    
    # One-Hot Encode specified columns
    df_processed = pd.get_dummies(df_processed, columns=one_hot_encode_cols, drop_first=False)

    return df_processed

# Example usage:
# Define the columns
label_encode_cols = ['loan_grade', 'cb_person_default_on_file']  # Columns to label encode
one_hot_encode_cols = ['person_home_ownership', 'loan_intent']  # Columns to one-hot encode

# Assuming `df` is your DataFrame
new_data = preprocess_categorical_data(new_data, label_encode_cols, one_hot_encode_cols)

```
